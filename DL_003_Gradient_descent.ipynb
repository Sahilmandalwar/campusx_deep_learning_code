```python
def perceptron_gradient_descent(X,y):
  w1 = w2 = b = 1
  lr = 0.01
  epoches = 100
  for epoch in range(epoches):
    for i in range(X.shape[0]):
      # find function value
      z = w1 * X[i][0] + w2 * X[i][1] + b

      if y[i] * z < 0 :
        w1 = w1 + lr * y[i] * X[i][0]
        w2 = w2 + lr * y[i] * X[i][1]
        b = b + lr * y[i]
  return w1,w2,b

```
