```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
AND_df = pd.DataFrame()
OR_df = pd.DataFrame()
EXOR_df = pd.DataFrame()
```

```python
AND_df['input 1'] = [0,1,0,1]
AND_df['input 2'] = [0,0,1,1]
AND_df['output'] = [0,0,0,1]
```

```python
OR_df['input 1'] = [0,0,1,1]
OR_df['input 2'] = [0,1,0,1]
OR_df['output'] = [0,1,1,1]
```

```python
EXOR_df['input 1'] = [0,1,0,1]
EXOR_df['input 2'] = [0,0,1,1]
EXOR_df['output'] = [0,1,1,0]
```

```python
sns.scatterplot(x=EXOR_df['input 1'],y=EXOR_df['input 2'], hue=EXOR_df['output'])
```

```python
from sklearn.linear_model import Perceptron
```

```python
clf1 = Perceptron()
clf2 = Perceptron()
clf3 = Perceptron()
```

```python
clf1.fit(AND_df[['input 1','input 2']].values,AND_df['output'].values)
clf2.fit(OR_df[['input 1','input 2']].values,OR_df['output'].values)
clf3.fit(EXOR_df[['input 1','input 2']].values,EXOR_df['output'].values)
```

```python
clf1.coef_ , clf1.intercept_
```

```python
clf2.coef_ , clf2.intercept_
```

```python
clf3.coef_ , clf3.intercept_
```

```python
x = np.linspace(-1,1,5)
y1 = -(clf1.coef_ [0][0] / clf1.coef_[0][1]) * x - (clf1.intercept_ / clf1.coef_[0][1])
y2 = -(clf2.coef_ [0][0] / clf2.coef_[0][1]) * x - (clf2.intercept_ / clf2.coef_[0][1])
y3 = -(clf3.coef_ [0][0] / clf3.coef_[0][1]) * x - (clf3.intercept_ / clf3.coef_[0][1])
```

```python
sns.scatterplot(x=OR_df['input 1'],y=OR_df['input 2'], hue=OR_df['output'])

plt.plot(x,y2,label='OR')

plt.legend()
```

NO CLASSIFICATION HERE!!

```python
plt.plot(x,y3,label='EXOR')
sns.scatterplot(x=EXOR_df['input 1'],y=EXOR_df['input 2'], hue=EXOR_df['output'])
plt.legend()
```

```python
plt.plot(x,y1,label='AND')
sns.scatterplot(x=AND_df['input 1'],y=AND_df['input 2'], hue=AND_df['output'])
plt.legend()
```
