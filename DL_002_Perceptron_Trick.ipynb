```python
import numpy as np
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100, n_features= 2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class = 1, random_state=41, hypercube=False, class_sep=5)
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1] , c=y, cmap='winter', s=100)
```

```python
def step(n):
  return 1 if n > 0 else 0
```

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)
  weight = np.ones(X.shape[1])
  lr = 0.1
  epochs = 1000
  for epoch in range(epochs):
    j = np.random.randint(X.shape[0])
    y_hat = step(np.dot(X[j],weight))
    weight = weight + lr*(y[j] - y_hat)*X[j]

  return weight[0], weight[1:]
```

```python
intercept_ , coeff_ = perceptron(X,y)
```

```python
intercept_, coeff_
```

```python
slope = -coeff_[0] / coeff_[1]
actual_intercept = -intercept_ / coeff_[1]
```

```python
plt.figure(figsize=(16,6))
plt.scatter(X[:,0], X[:,1] , c=y, cmap='winter', s=100)
plt.plot(X[:,0], slope*X[:,0] + actual_intercept)
plt.ylim(-3,2)
```
